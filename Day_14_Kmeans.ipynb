```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/Mall_Customers.csv")

x = data.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import KMeans

lst = []

for i in range(1,16):
  kmeans = KMeans(n_clusters = i , init = 'k-means++',random_state = 0)
  kmeans.fit(x)
  lst.append(kmeans.inertia_)

plt.plot(range(1,16),lst)
plt.title("K-Means Clustering")
plt.xlabel("Number of clusters")
plt.ylabel("Number of entries")
plt.show()
```

```python
kmeans = KMeans(n_clusters= 10, init = "k-means++")

y_kmeans = kmeans.fit_predict(x)

```

```python
plt.figure(figsize=(12,7))
plt.scatter(x[y_kmeans == 0,0],x[y_kmeans == 0,1,],s = 10, c= "red",label = "Cluster1")
plt.scatter(x[y_kmeans == 1,0],x[y_kmeans == 1,1,],s = 10, c= "blue",label = "Cluster2")
plt.scatter(x[y_kmeans == 2,0],x[y_kmeans == 2,1,],s = 10, c= "orange",label = "Cluster3")
plt.scatter(x[y_kmeans == 3,0],x[y_kmeans == 3,1,],s = 10, c= "green",label = "Cluster4")
plt.scatter(x[y_kmeans == 4,0],x[y_kmeans == 4,1,],s = 10, c= "yellow",label = "Cluster5")
plt.scatter(x[y_kmeans == 5,0],x[y_kmeans == 5,1,],s = 10, c= "purple",label = "Cluster6")
plt.scatter(x[y_kmeans == 6,0],x[y_kmeans == 6,1,],s = 10, c= "gold",label = "Cluster7")
plt.scatter(x[y_kmeans == 7,0],x[y_kmeans == 7,1,],s = 10, c= "pink",label = "Cluster8")
plt.scatter(x[y_kmeans == 8,0],x[y_kmeans == 8,1,],s = 10, c= "indigo",label = "Cluster9")
plt.scatter(x[y_kmeans == 9,0],x[y_kmeans == 9,1,],s = 10, c= "brown",label = "Cluster10")

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],c = "black", label = "Centeroid")
plt.title("K-Means final cluster")
plt.xlabel("Annual income")
plt.ylabel("spending score")
plt.legend()
plt.show()
```
