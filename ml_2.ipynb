```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset("iris")
```

```python
plt.plot(data['sepal_width'])
plt.xlabel("Indexes")
plt.ylabel("sepal width")

plt.show()
```

```python
plt.figure(figsize = (10,5))
plt.plot(data['sepal_width'], color = "red", label = "sepal width")
plt.plot(data['sepal_length'], color = "green", label = "sepal length")
plt.xlabel("Indexes")
plt.ylabel(" sepal counts")

plt.title("sepal graph", fontsize = 27)

plt.show()
```

```python
plt.plot(data['petal_width'], color = "red", label = "petal width")
plt.plot(data['petal_length'], color = "green", label = "petal length")
plt.xlabel("Indexes")
plt.ylabel(" petal counts")

plt.title("petal graph", fontsize = 27)

plt.show()
```

```python
plt.figure(figsize = (10,5))
plt.grid()
plt.plot(data['sepal_width'], color = "red", label = "sepal width", linestyle = "--")
plt.plot(data['sepal_length'], color = "green", label = "sepal length", marker = "o")
plt.xlabel("Indexes")
plt.ylabel(" sepal counts")

plt.title("sepal graph", fontsize = 27)

plt.show()
```

```python
plt.bar(data["species"],data["sepal_length"])
plt.bar(data["species"],data["sepal_width"])
plt.show()
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
tips.head()
```

```python
tips.describe()
```

```python
plt.bar(tips["day"],tips["total_bill"])
plt.show()
```

```python
plt.barh(tips["day"],tips["total_bill"])
plt.show()
```

```python
plt.hist(tips["total_bill"], edgecolor ="black")
plt.show()
```

```python
day = tips.groupby('day')["total_bill"].sum().reset_index()
day
```

```python
x = day["day"]

y = day["total_bill"] 
```

```python
explode = [0,0,0,0.1]
plt.pie(y,labels = x , explode = explode ,autopct='%1.1f%%',colors = ['red','green','blue','yellow'],
        wedgeprops = {'edgecolor': 'black'},shadow = True, startangle=90)
plt.show()
```

```python
plt.scatter(tips["total_bill"],tips["tip"])
plt.title("scatterplot")
plt.xlabel("total bill")
plt.ylabel("tip")

plt.show()
```

```python
tips['size'].value_counts()
```

```python
plt.scatter(tips["total_bill"],tips["tip"],s = 100 , c = tips['size'])
plt.title("scatterplot")
plt.xlabel("total bill")
plt.ylabel("tip")
cbar = plt.colorbar()
cbar.set_label("size")


plt.show()
```

```python
sns.distplot(tips["total_bill"])
```

```python

#Scatterplot using seaborn

sns.scatterplot(x = 'total_bill', y = 'tip', data = tips, hue = 'size', palette= 'summer')
```

```python
#Swarmplot

sns.swarmplot(x = 'sex', y = 'total_bill', data = tips)
```

```python
#Strip plot

sns.stripplot(x = 'day', y = 'tip', data = tips, hue = 'sex')
```

```python
#Voiline Plot

sns.violinplot(x = 'day', y = 'tip', data = tips, hue = 'sex', split = True)
```

```python
#Boxplot

sns.boxplot(x = 'day', y = 'tip', data = tips, hue = 'sex')
```
